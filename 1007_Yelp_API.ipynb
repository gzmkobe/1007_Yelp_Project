```python
def search_business(offset, category):
    url = 'https://api.yelp.com/v3/businesses/search'
    headers = {'Authorization': 'bearer %s' % access_token}
    params = {'location': 'New York City',
              'term': 'Restaurant',
              #'pricing_filter': '1, 2',
              'categories': category,
              #'sort_by': 'rating',
              'offset': str(offset),
              'limit': '50'
             }
    response = requests.get(url=url, params=params, headers=headers)
    return (response.json())['businesses']
```

```python
import numpy as np
def Search_result(category):
    offset_list=np.arange(0, 1000, 50)
    response=[]
    for offset in offset_list:
        response.append(pd.DataFrame(search_business(offset, category)))
    return pd.concat(response, ignore_index=True)
```

```python
def coordinates(dataset):
    result=[]
    for coordinate in dataset["coordinates"]:
        result.append((coordinate["latitude"], coordinate["longitude"]))
    return result
```

```python
def Plot_heat_map(fig, coordinate):
    fig.add_layer(gmaps.heatmap_layer(coordinate))
def Plot_symbol_map(fig, coordinate):
    starbucks_layer = gmaps.symbol_layer(coordinate, fill_color="green", stroke_color="green", scale=2)
    fig.add_layer(starbucks_layer)
def Plot_marker_map(fig, coordinate):
    markers = gmaps.marker_layer(coordinate)
    fig.add_layer(markers)
fig = gmaps.figure()
Plot_marker_map(fig, coordinates(search_result["dimsum"]))
fig
```
